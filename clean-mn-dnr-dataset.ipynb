# MN DNR
https://resources.gisdata.mn.gov/pub/gdrs/data/pub/us_mn_state_dnr/water_dnr_hydrography/metadata/dnr_hydrography_lakes_and_open_water.html

Downloaded ESRI File database from here (https://gisdata.mn.gov/dataset/water-dnr-hydrography) and loaded into Qgis then exported the feature table into an excel file and converted to csv. It might be possible to use python instead of Qgis gui to get feature table.

```python
df = pd.read_csv('mn-dnr-hydro-features.csv')
df.head()
```

I will select the features that we will use. Many of the features are related to drawing polygons on a map.

```python
features = [
    'fw_id', 'pw_basin_name', 'pw_parent_name', 'pw_sub_name',
    'sub_flag', 'wb_class', 'acres', 'shore_mi', 'cty_name'
]
df = df[features]
df.head()
```

Some of these features are not lakes. Let's get the count of each unique feature in the `wb_class` column. 

```python
df["wb_class"].value_counts()
```

So according to this dataset, there are 117,420 lakes or ponds in Minnesota. This is slightly less than the 124,400 reported by Stephen Aichele, a USGS geographer, in this Star Tribune Article. https://www.startribune.com/wisconsin-claims-it-has-more-lakes-than-minnesota-spoiler-alert-it-doesn-t/511709092/

This is only about 6% less, and Aichele referenced a different dataset. So overall, this number makes sense.

We still need to narrow down the data a bit more. The `sub_flag` column indicates that the record is a child of another record. This can be exemplified by search for "Minnetonka" in the `pw_basin_name` column.

```python
# select rows that are "Lake or Pond" only then drop wb_class
df = df[df["wb_class"] == "Lake or Pond"]
df = df.drop(["wb_class"], axis=1)

minnetonka = df[df["pw_basin_name"].str.contains("Minnetonka", na=False)].head()
minnetonka.head()
```

We can safely discard these sub records (bays, arms, etc.).

```python
# Select all where sub_flag is not Y - this includes many NaN
df = df[df["sub_flag"] != "Y"]
df.head()
```

We can see that there are a lot of NaN values.

```python
# number of NaN values in sub_flag
df['sub_flag'].isna().sum()
```

```python
# number of N values in sub_flag
len(df[df['sub_flag'] == 'N'])
```

This figure is similar to the number of lakes Finke got from the DNR. I want to explore these `NaN` values a bit more. I expect most of these are small and unnamed.

### sub_flag == NaN

#### Size

```python
sub_flag_na = df[df['sub_flag'].isna()]
sub_flag_na['acres'].describe()
```

The mean acreage is less than 1. This means the average lake in this subset is about 10% of the size necessary to be considered a lake.

#### Names

```python
sub_flag_na['pw_basin_name'].value_counts()
```

This shows that out of the 90,000+ lakes, only a few have names (including 'unnamed').

### sub_flag ==  N
#### Size

```python
sub_flag_notna = df[df['sub_flag'].notna()]['acres']
sub_flag_notna.describe()
```

The mean for lakes that difference is pretty significant. Now that we have removed the sub basins and characterized them we can remove the related features.

```python
df = df.drop(["pw_parent_name","pw_sub_name","sub_flag"], axis=1)
```

```python
# This returns the largest NaN lakes. The largest of these is actually not a lake!
# If you take the coordinates and view it in QGIS then find this same location on Google maps satellite view
# then you can see it is just a field! Interestingly, google maps regular view shows it as a lake as well.
# df[df['sub_flag'].isna()].sort_values('acres').tail()
```

When sorting by acreage we can see a few lakes that are accounted for twice. These large lakes are on the border of Minnesota. One row shows the portion in Minnesota, and the other shows the portion in the other regions. 

```python
df.sort_values(by=["acres"])
```

We could filter by `cty_name`, but the DNR documentation on attributes shows that the `fw_id` for non-Minnesotan portion of border waters will be `99999` and 'alternate' (duplicate) representations of basins are `88888`. So we can select everything with a `fw_id` less than `88888` (or `NaN`) to exclude both of these. https://resources.gisdata.mn.gov/pub/gdrs/data/pub/us_mn_state_dnr/water_dnr_hydrography/metadata/dnr_hydrography_lakes_and_open_water.html

```python
# select records by fw_id and drop feature
df = df[(df['fw_id'] < 88888) | (df['fw_id'].isna())]
df = df.drop(['fw_id'], axis=1)
len(df)
```

At this point we have a complete list of lakes and ponds. We have removed all other basin types, excluded some duplicate and out-of-bounds records, and removed sub basins. We can cleanup the headings, reindex and export a full version and lake version (records >= 10 acres).

```python
new_columns = {
    'pw_basin_name': 'name', 
    'shore_mi': 'shore_miles', 
    'cty_name': 'county'
}
df = df.rename(columns= new_columns)
```

```python
df = df.reset_index(drop=True)
df.to_csv('lakes-and-ponds.csv')
```

```python
lakes = df[df["acres"] >= 10]
lakes = lakes.reset_index(drop=True)
lakes.to_csv('lakes.csv')
```

This number is a bit larger than I would expect given that the Minnesota DNR officially recognizes 11,842 lakes.
